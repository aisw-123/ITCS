```python
import cv2
import numpy as np
import os
# Playing video from file:
cap = cv2.VideoCapture('13.mov')

try:
    if not os.path.exists('data'):
        os.makedirs('data')
except OSError:
    print ('Error: Creating directory of data')

currentFrame = 0
while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()
    if not ret : break

    # Saves image of the current frame in jpg file
    name = 'C:\\Users\\pmdna\\data\\frame' + str(currentFrame) + '.jpg'
    print ('Creating...' + name)
    cv2.imwrite(name, frame)

    # To stop duplicate images
    currentFrame += 1

# When everything done, release the capture
print (currentFrame)
cap.release()
cv2.destroyAllWindows()
```
